### 1.언어 기술

 인공지능의 언어 기술의 가장 기본적인 것은 인간의 언어를 컴퓨터에 입력하는 방법일 것이다. 가장 대표적인 예는 **자연어처리** 기법이다. 즉,인간이 사용하는 언어를 기계적으로 분석해 기계가 이해할 수 있는 형태로 만드는 것이다. 이러한 기술에 머신런닝 및 딥런닝 기술을 이용하여 컴퓨터는 문자의 형태와 감정, 문법을 빅데이터를 기반으로 분석하여 입력하고 그 도출값을 내는 기술이다.
 
 이 기술은 **금융,통신,번역** 등 비대면서비스를 하는 모든 분야에서 활발하게 서비스 및 개발 중 이다. 

 

####1-1.**카카오뱅크 상담 챗봇**


<img src="https://t1.daumcdn.net/cfile/tistory/999BF63E5B2C8B8F0B" width="30%" height="15%">

 
챗봇은 **채팅+로봇**의 합성어로 사용자의 질문을 해석해 최적의 답변을 찾아주는 서비스이다. 카카오뱅크의 상담 챗봇은 올해 6월에 시행 되었으며 **패턴인식**과 **머신러닝 기술**을 통한 자연어인식 기술이 사용된다.
 
  현제는 키워드를 통한 상담내용 분류와 상품 안내서비스를 진행하고 있으며 채팅뿐만 아니라 음성인식을 통한 서비스도 제공 되고 있다. 
  
   앞으로는 머신런닝 기술을 이용하여 각 개별 사용자들의 챗봇이용 형태와 패턴을 분석하여 고객의 필요에 맞는 서비스를 제공할 예정이다. 예를 들면 고객의 현제 운용자산과 구매상품의 특정 한도를 분석하여 필요한 금융상품을 소개해주고 판매도 하며 다양한 업무를 진행 할 예정이다. 

 한때 금융업에 종사하면서 느낀 점은 다양한 상품을 정리하여 필요한 고객에게 알려주는 인프라 및 정보 분류가 되지 않아 판매에 어려움을 격은 적이 있었다. 그러나 인공지능을 통해 손님의 니즈에 맞는 상품을 구별하고 추천해주는 방식은 그 산업의 매출 증대에 지대한 영향을 끼칠 것으로 보인다. 

   또한 현제 출시되는 금융상품 중 인공지능이 운용하는 상품이 있다는 것에서도 많은 귀감을 주고 있다.

### **2.음성기술**

 음성인식 기술은 컴퓨터가 인간의 말을 알아듣는 기술로서 50년 이상의 연구개발 역사를 가지고 있다. 그간 꾸준한 기술개발과 상용화 시도를 통하여 지속적인 발전을 이루어왔으며 최근 들어 스마트폰 활성화에 따라 관심도가 급속히 높아지고 있는 분야이다.
현재 구글, 애플, 마이크로소프트 등 세계적인 IT 선도 기업이 음성인식 기술 개발에 많은 노력을 기울이고 있으며 특히 스마트폰 환경에서의 음성인식 응용 서비스인 음성검색, 자동통역, 인공지능 개인비서 등을 시작하며 본격적인 기술, 시장 선점 경쟁에 돌입하였다.

#### **2-2. 인공지능 스피커**
 인공지능 비서와 같이 가장 가까운 음성인식 인공지능은 인공지능 스피커이지 않을 까 싶다. SKT "아리야, KT "기가지니", 카카오 "카카오미니, 네이버 "클로버" 및 해외 기업인 아마존 "에코", 구글 "홈", 에플 "홈팟" 등 세계 각국에서 음성인식 기반 인공지능을 서로 출시하고 있다. 터치와 버튼이 아닌 음성인식이 기계를 제어하는 가장 기초적인 형태라고 할 수 있겠다. 


 공통적으로 이 스피커들은 연결되있는 도메인에 기반을 둔 음성기능은 잘 소화 한다는 것이다. 예를 들면 **음악, 뉴스, 날씨, 일정관리,알람을 들 수 있다. 노래의 곡명 혹은 가수 검색을 기반으로 연동된 도메인 서비스를 제공하는 기능은 뛰어 나다. 그러나 자연어 해석능력의 발달로 인해 음식배달, 카카오택시 등 다양한 서비스도 제공 되고 있고 더욱더 발전할 것이다.

 인공지능 스피커를 허브로 만들어 스마트 허브 기능으로 발전 되고 있다. 전자기기에 사물인터넷(IoT)이 장착되어 인공지능 스피커를 통해 가정 내 모든 기기를 음성으로 제어하는 것이다. 대표적으로 LG "스마트씽큐 허브", 애플 "홈팟" 등을 예로들수 있다. "에어컨 켜줘, 세탁기 몇 분 남았어? "등 다양한 명령을 통하여 가전제품을 사용 할 수 있게 된 것이다.

### **3.이미지 기술**

 인공지능 이미지 기술은 기존 컴퓨터 비전을 중심으로 연구되어온 분야 이었다. 거기에 인공신경망기술인 딥런닝을 통해 비약적인 발전을 하여 등장한 것이다. 이미지를 여러 패턴으로 분류해서, 패턴형 데이터를 학습하여 새로운 이미지가 주어질 때 이것이 무엇인지 정확하게 알아내는 원리이다.

### **4.자율주행 기술**

 자율주행 기술은 모든 인공지능 기술의 복합체라고 볼 수 있다. 머신런닝,자연어처리,이미지처리,음성인식 모듈을 통해 자율적으로 주행 할 수 있게 만든 기술이기 때문이다. 예를 들면 "우리 집까지 운전해줘."라는 명령어를 입력하면 인공지능은 집까지 최단 거리를 계산하고 출발을 하며 주변 교통 환경에 맞춰 멈추고 출발하는데 이때 시야확보를 하는 센서 모듈이 작동하게 되는 복합적인 기술인 것이다.

 #### **4-1.자율주행 자동차**

  자율주행 자동차는 운전자 개입 없이 자동차가 스스로 도로 환경을 인식하고 조향, 변속장치 등을 제어하여 목적지까지 주행하는 것을 말합니다. 자율주행차의 구조는 센서, 프로세서, 알고리즘, 액추에이터로 구성되는데 센서가 데이터를 수집하면 프로세스가 데이터를 처리하고 알고리즘이 데이터를 해석, 주행경로 및 속도 등 운행에 대해 결정을 내리면 액추에이터가 운행에 필요한 시스템을 제어 하는 방식이다.
 이러한 무수한 데이터를 모의기 위해 기계학습이 필요한데 이때 필요한 기술이 딥러닝이다. 구글에서는 2017년 기준 약700백 킬로를 주행하며 빅데이터를 모았다고 시사 한바가 있다.
 자율주행 자동차는 사실 자율주행 시장의 일부일 뿐이다. 인간이 이용하는 모든 교통수단에 적용 가능하며 어떠한 기술이 나올지는 앞으로 귀추가 주목되고 있다.
